**C1.** 위에 데이터를 가지고 user_id가 NaN인 행은 모두 지우고, platform과 browser는 최빈값으로 대체하시오. (2점)

```python
info_login = info_login.iloc[info_login['user_id'].dropna().index,:]
info_login = pd.DataFrame(info_login)

most_platform = info_login['platform'].mode()[0]
info_login['platform'] = info_login['platform'].fillna(most_platform)

most_browser = info_login['browser'].mode()[0]
info_login['browser'] = info_login['browser'].fillna(most_browser)
```

**C4.** 각 플랫폼 별 user_id의 평균 값을 딕셔너리에 넣어주는 코드를 만들어보세요. (3점)
- key는 플랫폼 명칭, value는 user_id의 평균으로 해주세요.
- 3주차에서 배운 사용자 정의 함수 (boss_question)를 만들어서 사용해보세요. 
- def boss_question(x):
- 함수를 통한 리스트를 map을 이용하여 반올림하시오.
- 힌트) def, query, for, unique 사용

```python
dict={}
def boss_question(x):
    m=int(info_login.query("platform==@x")["user_id"].mean())
    dict[x]=m
platforms=info_login["platform"].unique()
for a in platforms:
    boss_question(a)

print(dict)
```
